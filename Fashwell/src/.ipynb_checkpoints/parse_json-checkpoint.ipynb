```python
def get_resolution(row,folder='../ImagesFashwell/'):
    name = row.image_id+'.jpg'
    a = os.popen('identify '+folder+name)
    result = a.read()
    a.close()
    w,h=result.split()[2].split('x')
    return pd.Series({'w':int(w),'h':int(h)})
```

```python
df[['w','h']]=df.apply(get_resolution,axis=1)
```

```python
colours = ['red','blue','green','orange','black']
```

```python
def create_output(group,folder='./ImagesFashwell/'):
    filename=folder+group['image_id'].values[0]
    pipe = 'convert '+filename+'.jpg'+" -fill none -strokewidth 30"
#     print('ok')
    temp = []
    for index,row in group.iterrows():
        w=row.w
        h=row.h
        colour = colours[row.colour]
        startx = round(row['bbox'][0]*w)
        starty = round(row['bbox'][1]*h)
        endx = round(row['bbox'][2]*w)+startx
        endy = round(row['bbox'][3]*h)+starty
        pipe+=(' -stroke '+colour+' -draw "rectangle '+str(startx)+','+str(starty)+' '+str(endx)+','+str(endy)+'"')
        p = " ".join(row['product'].split('/')[-1].split('-')[0:-2]).title()
        temp.append({'rect':[startx, startx, endx, endy],'colour':colour,'product':p,'link':row["product"],'category':row.category_id})
    pipe+=" "+filename+'BOX.jpg'
    a = os.popen(pipe)
    result = a.read()
    a.close()
    return pd.Series({'command':pipe,'outputs':temp})
```

```python
df2 = df.groupby('image_id').apply(create_output)
```

```python
def create_html(group,folder="./ImagesFashwell/"):
    name=" ".join(group.image_id.values[0].split('-')[0:2]).title()
    html=''
    html+='<h1>'+name+'</h1>'
    html+='<div class="output" id="'+group.image_id.values[0]+'"><img class="img_output" src="'+folder+group.output.values[0]+'"></div>'
    html+='<div class="products"><ul>'
    for index,row in group.iterrows():
        p = " ".join(row['product'].split('/')[-1].split('-')[0:-2]).title()
        html+='<li style="color:'+row.colour_box+'"><span style="color:'+row.colour_box+'">Category: '+row.category_id+'. <a style="color:'+row.colour_box+'" target="_blank" href="'+row["product"]+'">'+'Best match: '+p+'</a></span></li>'
    html+='</ul></div>'
    return pd.Series({'html':html})
```
